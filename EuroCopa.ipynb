Extracci√≥n de los datos de la carpeta events

```python
# Ruta de la carpeta
events = '/content/drive/MyDrive/Eurocopa/DATOS/events'
```

```python
# Lista para guardar los archivos disponibles
lista_archivos_events = []

# Recorre todos los archivos en la carpeta
for archivo in os.listdir(events):

    lista_archivos_events.append(archivo)
```

```python
# Todos los archivos disponibles
lista_archivos_events
```

```python
# Seleccionamos a los archivos que sean de la eurocopa
archivos_eurocopa_events = [archivo for archivo in lista_archivos_events if 'eurocopa' in archivo]
```

```python
# Archivos solo de la eurocopa
archivos_eurocopa_events
```

```python
# Lista para almacenar los dataframes
lista_df = []

# Recorre todos los archivos en la carpeta
for archivo in os.listdir(events):
    # Verifica que el archivo sea un .csv y est√© en la lista de archivos a importar
    if archivo.endswith('.csv') and archivo in archivos_eurocopa_events:
        ruta_archivo = os.path.join(events, archivo)
        # Lee el archivo y lo a√±ade a la lista
        df = pd.read_csv(ruta_archivo)
        lista_df.append(df)

# Concatena todos los dataframes en uno solo
df_final_events = pd.concat(lista_df, ignore_index=True)
```

```python
df_final_events.head()
```

```python
√á# ID de los equipos de la euro

# Elimina las filas duplicadas bas√°ndote en 'Team ID' y 'Team Name'
df_unicos = df_final_events.drop_duplicates(subset=['Team ID', 'Team Name'])


print('ID de las selecciones de la eurocopa')
print('----'*10)
# Ahora itera sobre el nuevo DataFrame
for i, fila in df_unicos.iterrows():
    print(f"Team ID: {fila['Team ID']}, Team Name: {fila['Team Name']}")
```

```python
# Cambiar la ruta de la carpeta a la correspoda
teams = '/content/drive/MyDrive/Eurocopa/DATOS/teams'
```

```python
# Lista para guardar los archivos
lista_archivos_teams = []

# Recorre todos los archivos en la carpeta
for archivo in os.listdir(teams):

    lista_archivos_teams.append(archivo)
```

```python
lista_archivos_teams
```

```python
# Si los archivos coinciden con eurocopa se guardan
archivos_eurocopa_teams = [archivo for archivo in lista_archivos_teams if 'eurocopa' in archivo]
```

```python
# Archivos de la eurocopa
archivos_eurocopa_teams
```

```python
# Lista para almacenar los dataframes
lista_df = []

# Recorre todos los archivos en la carpeta
for archivo in os.listdir(teams):
    # Verifica que el archivo sea un .csv y est√© en la lista de archivos a importar
    if archivo.endswith('.csv') and archivo in archivos_eurocopa_teams:
        ruta_archivo = os.path.join(teams, archivo)
        # Lee el archivo y lo a√±ade a la lista
        df = pd.read_csv(ruta_archivo)
        lista_df.append(df)

# Concatena todos los dataframes en uno solo
df_teams = pd.concat(lista_df, ignore_index=True)
```
